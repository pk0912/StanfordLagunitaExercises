```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
```

```python
df = pd.read_excel('../Datasets/friends.xls')
```

```python
labels = df['Friends'].value_counts().index
```

```python
data = df['Friends'].value_counts().values
```

```python
plt.pie(df['Friends'].value_counts().values,labels=df['Friends'].value_counts().index,autopct='%1.2f%%')
plt.show()
```

```python
the_grid = GridSpec(2, 1)
explode = [0.3,0.2,0]
plt.subplot(the_grid[0, 0], aspect=1)

plt.pie(data, labels=labels, autopct='%1.1f%%', shadow=True, radius=5, wedgeprops = {'linewidth': 2}, center= (0,0))

plt.show()
plt.subplot(the_grid[1, 0], aspect=1)

plt.pie(data, explode=explode, labels=labels, autopct='%.0f%%', shadow=True, radius=5, center= (0,20))

plt.show()
```
