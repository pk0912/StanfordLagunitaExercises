```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data = pd.read_excel('../Datasets/olympics_2012.xls')
```

```python
#Scatter Plot
plt.scatter(x=data['Year'].values,y=data['Time'].values)
plt.plot()
```

```python
sns.lmplot(data=data, x='Year', y='Time', fit_reg=True)
plt.plot()
```

```python
#Pairplot
sns.pairplot(data=data, x_vars='Year', y_vars='Time', kind='reg', size=5)
plt.plot()
```

## Linear Regression

```python
X = data['Year'].values
Y = data["Time"].values
X = X.reshape(27,1)
Y = Y.reshape(27,1)
```

```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(X,Y)
```

```python
lin_reg_model.intercept_
```

```python
lin_reg_model.coef_
```

```python
X.shape,Y.shape
```

## Removing the outlier and calculate least square regression line's equation again

```python
data2 = pd.read_excel('../Datasets/olympics_2012.xls')
```

```python
data2 = data2.drop(data2.loc[data2['Time']==data2['Time'].values.max()].index[0])
```

```python
data2 = data2.reset_index()
```

```python
data2 = data2.drop('index',axis=1)
```

```python
x = data2['Year'].values
x = x.reshape(26,1)
y = data2['Time'].values
y = y.reshape(26,1)
```

```python
lin_reg_model2 = LinearRegression()
lin_reg_model2.fit(x,y)
```

```python
lin_reg_model2.coef_
```

```python
lin_reg_model2.intercept_
```

```python
x.shape , y.shape
```
