```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats.stats import pearsonr
```

```python
data = pd.read_excel('../Datasets/animals.xls')
```

```python
facet = sns.lmplot(data=data, x='longevity', y='gestation', hue='animal',\
                   fit_reg=False, legend=True, legend_out=True)
```

## Scatter Plot

```python
plt.scatter(x=data['longevity'].values,y=data['gestation'].values)
plt.plot()
```

## Finding correlation

```python
np.correlate(data['longevity'].values,data['gestation'].values)
```

```python
# np.corrcoef returns correlation matrix
cor_mat = np.corrcoef(data['longevity'].values,data['gestation'].values)
```

```python
print("Correlation :",cor_mat[0][1])
```

```python
# Calculate a Pearson correlation coefficient and the p-value for testing non-correlation.
pearson_correlation_coefficient = pearsonr(data['longevity'].values,data['gestation'].values)
```

```python
pearson_correlation_coefficient
```

## Remove the outlier and then again calculate the correlation

```python
data = data.drop(data.loc[data['gestation']==data['gestation'].values.max()].index[0])
```

```python
data = data.reset_index()
```

```python
data = data.drop('index',axis=1)
```

```python
plt.scatter(x=data['longevity'].values,y=data['gestation'].values)
plt.plot()
```

```python
## measuring correlation coefficient
np.corrcoef(data['longevity'].values,data['gestation'].values)[0,1]
```
