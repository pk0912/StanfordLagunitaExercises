```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('../Datasets/height.xls')
```

```python
plt.scatter(x=data['height'].values,y=data['weight'].values)
plt.plot()
```

## plotting a labelled scatter plot

```python
data.loc[data['gender']==0,'gender'] = 'male'
data.loc[data['gender']==1,'gender'] = 'female'
```

```python
facet = sns.lmplot(data=data, x='height', y='weight', hue='gender',\
                   fit_reg=False, legend=True, legend_out=True)
```

```python
sns.lmplot(data=data, x='height', y='weight', hue='gender',\
                   fit_reg=True, legend=True, legend_out=True)
plt.plot()
```
